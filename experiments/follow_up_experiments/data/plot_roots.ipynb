```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
forewing = np.genfromtxt('hercules_forewing_n=20.csv', delimiter=',')
circle = np.genfromtxt('circle_n=20_d=2.csv', delimiter=',')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6), sharex=True, sharey=True)

num_ticks = 5
x_ticks = np.linspace(-1, 1, num_ticks)
y_ticks = np.linspace(-1, 1, num_ticks)

for ax in axes:
    ax.set_xticks(x_ticks)
    ax.set_yticks(y_ticks)
    ax.tick_params(axis='both', labelsize=16)

axes[0].plot(forewing[0::2], forewing[1::2], 'o', color='steelblue', markersize=10)
axes[0].set_title('Forewing', size=20)
axes[0].set_aspect('equal')
axes[0].tick_params(axis='both', labelsize=16)

axes[1].plot(circle[0::2], circle[1::2], 'o', color='steelblue', markersize=10)
axes[1].set_title('Circle', size=20)
axes[1].set_aspect('equal')
axes[1].tick_params(axis='both', labelsize=16)

plt.tight_layout()
plt.savefig("sim_root_shapes.pdf")
plt.show()
```
